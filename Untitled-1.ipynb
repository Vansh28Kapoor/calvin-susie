```python
import tensorflow as tf

filename = "augmented_dataset/susie_augmented_0.tfrecord"

PROTO_TYPE_SPEC = {
    "actions": tf.float32,
    "proprioceptive_states": tf.float32,
    "image_states": tf.uint8,
    "language_annotation": tf.string,   # plain text
}
extended_spec = {**PROTO_TYPE_SPEC, "susie_goal_images": tf.uint8}

# All features were stored as bytes/strings in the Example
features = {k: tf.io.FixedLenFeature([], tf.string) for k in extended_spec.keys()}

raw_ds = tf.data.TFRecordDataset([filename])

def _parse_fn(serialized_example):
    ex = tf.io.parse_single_example(serialized_example, features)
    out = {}
    for k, dtype in extended_spec.items():
        if dtype is tf.string:
            # already a scalar tf.string Feature; don't parse as TensorProto
            out[k] = ex[k]
        else:
            # these must have been written with tf.io.serialize_tensor(...)
            out[k] = tf.io.parse_tensor(ex[k], out_type=dtype)
    return out

ds = (raw_ds
      .map(_parse_fn, num_parallel_calls=tf.data.AUTOTUNE)
      .prefetch(tf.data.AUTOTUNE))

# quick sanity check
for sample in ds:
    for k, v in sample.items():
        print(k, v.shape, v.dtype)
```
